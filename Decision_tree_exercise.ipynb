```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
df=pd.read_csv('D:\\internship\\covid_vaccine_statewise.csv')
df.head()
```

```python
new_df=df[['Updated_On','Total_Sessions_Conducted','First_Dose_Administered','Second_Dose_Administered','Total_Individuals_Vaccinated']]
new_df.head()
```

```python
plt.scatter(df.Total_Individuals_Vaccinated,df.Updated_On,marker="+",color="red")
```

```python
inputs=new_df.drop('Total_Individuals_Vaccinated',axis='columns')
target=df.Total_Individuals_Vaccinated
```

```python
inputs.First_Dose_Administered = inputs.First_Dose_Administered.fillna(inputs.First_Dose_Administered.mean())
inputs.Second_Dose_Administered = inputs.Second_Dose_Administered.fillna(inputs.Second_Dose_Administered.mean())
#inputs.First_Dose_Administered[:10]
# now i need to remove "NaN" in that place i will fill it with mean age
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()

```

```python
# Now train model between them
model.fit(X_train,y_train)
```
